### Data Cleaning
#### 1)Checked missing values
#### 2)Filled missing values with "Unknown"
#### 3)Removed duplicate records

```python
df.isnull().sum()
```

```python
# fill missing values (simple beginner approach)
df.fillna("Unknown", inplace=True)
```

```python
# check duplicates
print(df.duplicated().sum())
```

```python
df['City Name'].replace('Unknown', np.nan, inplace=True)
df['City Name'].fillna('Not Specified', inplace=True)
```

### 1) Top 10 Cities with Most Accidents
#### This graph shows which cities have the highest number of accidents.

```python
top_cities = df['City Name'].value_counts().head(10)

plt.figure()
sns.barplot(x=top_cities.values, y=top_cities.index, color= "blue")
plt.title("Top 10 Cities with Most Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("City")
plt.show()
```

### 2) Accidents by Day of Week
#### Helps identify which days are more accident-prone.

```python
day_counts = df['Day of Week'].value_counts()

plt.figure()
sns.barplot(x=day_counts.index, y=day_counts.values, palette="plasma")
plt.xticks(rotation=45)
plt.title("Accidents by Day of Week")
plt.show()
```

### 3) Accident Severity Distribution (Pie Chart)
#### Shows percentage of minor, major, and fatal accidents.

```python
severity_counts = df['Accident Severity'].value_counts()

plt.figure()
plt.pie(severity_counts.values, labels=severity_counts.index, autopct='%1.1f%%')
plt.title("Accident Severity Distribution")
plt.show()
```

### 4) Age Group vs Accidents
#### Shows which age group is more involved in accidents.

```python
plt.figure()
sns.histplot(df['Driver Age'], bins=20, color='red')
plt.title("Driver Age Distribution")
plt.show()
```

### 5) Weather Conditions vs Accidents
#### Shows how weather affects accidents.

```python
weather_counts = df['Weather Conditions'].value_counts()

plt.figure()
sns.barplot(x=weather_counts.index, y=weather_counts.values, palette="plasma")
plt.xticks(rotation=45)
plt.title("Accidents by Weather Conditions")
plt.show()
```

### 6) Vehicle Type Involved
#### Identifies which vehicle types are most involved.

```python
vehicle_counts = df['Vehicle Type Involved'].value_counts().head(10)

plt.figure()
sns.barplot(x=vehicle_counts.values, y=vehicle_counts.index, palette="rainbow")
plt.title("Vehicle Types Involved in Accidents")
plt.show()
```

### 7) Alcohol Involvement:
#### Shows impact of alcohol on accidents.

```python
alcohol_counts = df['Alcohol Involvement'].value_counts()

plt.figure()
sns.barplot(x=alcohol_counts.index, y=alcohol_counts.values,palette="viridis")
plt.title("Alcohol Involvement in Accidents")
plt.show()
```

### 8) Correlation Heatmap:
#### Shows relationship between numerical features like age, speed, casualties.
#### This heatmap shows the correlation between numerical variables in the dataset. It helps to understand how strongly different features are related to each other.

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure()
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
